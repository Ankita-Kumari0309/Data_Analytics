### Find the car (or cars) with the best ratio of horsepower to weight among all cars that also have above-median mpg.

```python
#your code here
median_mpg = Cars['mpg'].median()
above_median_cars = Cars[Cars['mpg'] > median_mpg].copy()
above_median_cars['hp_weight_ratio'] = above_median_cars['horsepower'] / above_median_cars['weight']
max_ratio = above_median_cars['hp_weight_ratio'].max()
best_ratio_cars = above_median_cars[above_median_cars['hp_weight_ratio'] == max_ratio]
print(best_ratio_cars[['name', 'hp_weight_ratio']])
```

### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

```python
#your code here
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate average mpg per year and origin
avg_mpg = Cars.groupby(['model_year', 'origin'])['mpg'].mean().reset_index()

# Set the plot style
sns.set(style="whitegrid")

# Create a lineplot with seaborn
plt.figure(figsize=(12, 6))
sns.lineplot(data=avg_mpg, x='model_year', y='mpg', hue='origin', marker='o')

plt.title('Evolution of Average MPG Over the Years by Origin')
plt.xlabel('model_year')
plt.ylabel('Average MPG')
plt.legend(title='Origin')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Create a Seaborn scatterplot (or PairGrid) where:

### We define a “consistent” car model as one that was produced over multiple years and had very low variation in mpg across those years (standard deviation < 1.0).

Tasks:

Identify car names that appear in more than one model_year.

For each such name, compute the standard deviation of mpg across years.

Return the car(s) with the lowest variation in mpg, among those with at least 2 appearances and std(mpg) < 1.0.

Report the model name(s), number of appearances, and the average mpg.

Bonus:
Sort the result by number of appearances (descending), then mpg (descending).

```python
#your code here
import pandas as pd

# Step 1: Finding the  car names that appear in more than one model_year
name_year_counts = Cars.groupby('name')['model_year'].nunique()
multi_year_names = name_year_counts[name_year_counts > 1].index

# Step 2: Filterring  the DataFrame for those names
multi_year_cars = Cars[Cars['name'].isin(multi_year_names)]

# Step 3: Compute std deviation, count of appearances, and average mpg for each name
stats = multi_year_cars.groupby('name').agg(
    mpg_std=('mpg', 'std'),
    appearances=('model_year', 'count'),
    avg_mpg=('mpg', 'mean')
).reset_index()

# Step 4: Filter for std(mpg) < 1.0
consistent_cars = stats[(stats['mpg_std'] < 1.0)]

# Step 5: Find the lowest variation std among consistent cars
min_std = consistent_cars['mpg_std'].min()
lowest_variation_cars = consistent_cars[consistent_cars['mpg_std'] == min_std]

# Bonus: Sort by appearances (descending) then avg_mpg (descending)
result = lowest_variation_cars.sort_values(by=['appearances', 'avg_mpg'], ascending=[False, False])

# Display the result with relevant columns
print(result[['name', 'appearances', 'avg_mpg']])
```
